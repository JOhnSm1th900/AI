```python
# دالة لإنشاء فلم
from moviepy.editor import ImageSequenceClip
```

```python
# قائمة لحفظ الصور المتتالية
frames = []
```

```python
# عدد الحلقات التجريبية
Episodes=10
```

```python
# إجمالي مكافآت كل الحلقات
total_reward_episode=0
```

```python
for episode in range(Episodes):
    # مؤشر الانتهاء
    done = False
    # إجمالي مكافآت الحلقة الواحدة
    total_reward = 0
    # إعادة التهيئة
    obs, info = env.reset()
    # إضافة صورة البيئة
    frames.append(env.render())
    while not done:
        # فعل عشوائي
        action = env.action_space.sample()
        # تنفيذ الفعل
        obs, reward, terminated, truncated, info = env.step(action)
        # إضافة صورة البيئة
        frames.append(env.render())
        # إضافة المكافأة الحالية لإجمالي المكافآت
        total_reward = total_reward + reward
        if terminated or truncated:
          done=True
    print(f"Episode: {episode + 1} - Total Reward: {total_reward}")
    total_reward_episode+=total_reward
```

```python
print(f"Average episode reward: {total_reward_episode/Episodes}")
```

```python
# إنشاء فلم من قائمة الصور
clip = ImageSequenceClip(frames, fps=10)
```

```python
# إغلاق البيئة
env.close()
```
