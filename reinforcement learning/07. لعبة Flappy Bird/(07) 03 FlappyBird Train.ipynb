```python
# استيراد المكتبات اللازمة
import gymnasium as gym
import flappy_bird_gymnasium
```

```python
# صنف التعلم المعزز
from stable_baselines3 import PPO
```

```python
# تضمين صنف التتبع
from stable_baselines3.common.monitor import Monitor
```

```python
# تضمين صنف التعامل مع عدة بيئات
from stable_baselines3.common.vec_env import DummyVecEnv
```

```python
# دالة التقييم
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# إنشاء البيئة
env = gym.make("FlappyBird-v0", render_mode="rgb_array", use_lidar=True)
```

```python
# استخدام صنف التتبع
env_wrapped = Monitor(env)
# استخدام صنف تعدد البيئات
env_wrapped = DummyVecEnv([lambda: env_wrapped])
```

```python
# التصريح عن نموذج التعلم المعزز
model = PPO("MlpPolicy", env_wrapped, verbose=0)
# عدد دورات التدريب
total_timesteps=200000
# التدريب
model.learn(total_timesteps=total_timesteps)
```

```python
# عدد حلقات التقييم
n_eval_episodes=32
```

```python
# تقييم النموذج
mean_reward, std_reward = evaluate_policy(model, env_wrapped, n_eval_episodes=n_eval_episodes)
print(f"Mean reward: {mean_reward:.2f} +/- {std_reward:.2f}")
```

```python
# مسار مجلد العمل
working_folder='/content/drive/MyDrive/RLModels/'
# اسم الملف
model_file_name=f'ppo_flappybird_RL_model_{total_timesteps}'
# مسار الملف
model_file_name_path= working_folder + model_file_name
```

```python
# حفظ النموذج المدرب
model.save(model_file_name_path)
```
