```python
import gymnasium as gym
# مكتبة التعامل مع الروبوتات
import panda_gym
# إنشاء بيئة العمل
env_id="PandaReach-v3"
#env_id="PandaPush-v3"
#env_id="PandaSlide-v3"
#env_id="PandaStack-v3"

env = gym.make(env_id, render_mode="rgb_array")
```

```python
# صنف التعلم المعزز
from stable_baselines3 import PPO
# تضمين صنف التتبع
from stable_baselines3.common.monitor import Monitor
# تضمين صنف التعامل مع عدة بيئات
from stable_baselines3.common.vec_env import DummyVecEnv
# دالة التقييم
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# استخدام صنف التتبع
env_wrapped = Monitor(env)
```

```python
# استخدام صنف تعدد البيئات
env_wrapped = DummyVecEnv([lambda: env_wrapped])
```

```python
# إنشاء نمو1ج التعلم
model = PPO("MultiInputPolicy", env_wrapped, verbose=0)
```

```python
# تدريب النموذج
print("Training the model...")
# عدد دورات التدريب
total_timesteps=50000
# التدريب
model.learn(total_timesteps=total_timesteps)
print("Training finished.")
```

```python
# عدد حلقات التقييم
n_eval_episodes=100
```

```python
# قياس الأداء
mean_reward, std_reward = evaluate_policy(model, env_wrapped, n_eval_episodes=n_eval_episodes)
print(f"Mean reward: {mean_reward} +/- {std_reward}")
```

```python
# مسار مجلد العمل
working_folder='/content/drive/MyDrive/RLModels/'
# اسم النموذج المدرب
model_file_name=f"ppo_{env_id}_{total_timesteps}"
# مسار ملف النموذج المدرب
model_file_path = working_folder + model_file_name
model.save(model_file_path)
print("model saved .")
```
