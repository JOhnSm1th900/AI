```python
# استيراد المكتبة
import gymnasium as gym

# تهيئة البيئة
env = gym.make("Taxi-v3", render_mode="rgb_array")

# إعادة تهيئة البيئة للحصول على الحالة الأولية
observation, info = env.reset()

print(observation)
```

```python
# إظهار الحالة الأولية
env.render()
```

```python
# تهيئة البيئة
env = gym.make("Taxi-v3", render_mode="ansi")
# إعادة تهيئة البيئة للحصول على الحالة الأولية
observation, info = env.reset()
# إظهار الحالة الأولية
grid=env.render()
print(grid)
```

```python
# طباعة عدد الأفعال و الحالات
print("State Space ", env.observation_space.n)
print("Action Space ", env.action_space.n)
```

```python
# تهيئة البيئة
env = gym.make("Taxi-v3", render_mode="ansi")
# تحديد عدد دورات الحلقة
env = gym.wrappers.TimeLimit(env, max_episode_steps=2)
# إعادة تهيئة البيئة للحصول على الحالة الأولية
observation, info = env.reset()
grid=env.render()
i=1
print(i)
print(grid)
```

```python
terminated=False
truncated=False
# إجراء حلقة للتفاعل مع البيئة
while not (terminated or truncated):
     # اختيار فعل عشوائي
    action = env.action_space.sample()
     # تنفيذ الفعل في البيئة
    next_state, reward, terminated, truncated, info = env.step(action)
    # عرض البيئة بعد كل خطوة
    grid=env.render()
    i=i+1
    print(f"step: {i}, next_state: {next_state}, reward: {reward}")
    print(grid)

# إغلاق البيئة بعد الانتهاء
env.close()
```
