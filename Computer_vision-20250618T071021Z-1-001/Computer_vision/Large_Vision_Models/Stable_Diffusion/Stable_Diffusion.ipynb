```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "dreamlike-art/dreamlike-photoreal-2.0"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompts = ["Twinkle, Twinkle, Mad Star",
          "there is no life under the mad star",
          "and ALL in Madness"]
```

```python
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    image.save('/content/drive/MyDrive/Computer_vision/Large_Vision_Models/Stable_Diffusion/' + f'picture_{i}.jpg')
```
