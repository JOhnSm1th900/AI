```python
import tensorflow as tf

x = tf.constant([1, 2, 3, 4], dtype=tf.float32)
y = tf.constant([2, 4, 6, 8], dtype=tf.float32)

W = tf.Variable([0.4], dtype=tf.float32)

b = tf.Variable([-0.4], dtype=tf.float32)

linear_model = W * x + b

error = linear_model - y

squared_errors = tf.square(error)

loss = tf.reduce_sum(squared_errors)

loss_value = loss.numpy()
print("Loss:", loss_value)
```

```python
learning_rate=0.01

optimizer = tf.optimizers.SGD(learning_rate=learning_rate)


epochs=1000

for i in range(epochs):
    with tf.GradientTape() as tape:
        linear_model = W * x + b

        error = linear_model - y

        squared_errors = tf.square(error)

        loss = tf.reduce_sum(squared_errors)
        print(loss)

    gradients = tape.gradient(loss, [W, b])

    optimizer.apply_gradients(zip(gradients, [W, b]))

W_result=W.numpy()[0]
b_result=b.numpy()[0]

print("Updated values of W and b:", round(W_result,2), round(b_result,2))
```
