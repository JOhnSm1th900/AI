```python
from keras.layers import Input

import numpy as np
from keras.models import Sequential
from keras.layers import Dense

x_train = np.array([1, 2, 3, 4])
y_train = np.array([2, 4, 6, 8])

model = Sequential()

model.add(Dense(units=1, activation='linear', input_shape=(1,)))

model.compile(optimizer='sgd', loss='mean_absolute_error', metrics=['mae', 'mape'])

epochs = 500
batch_size = 2

model.fit(x=x_train, y=y_train, epochs=epochs, batch_size=batch_size)
```

```python
weights, bias = model.get_weights()

print("Updated values of W and b:", round(weights[0][0],1), round(bias[0],1))
```

```python
import numpy as np

pred = model.predict(np.array([[5], [6], [7]]))
print(np.round(pred))
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

# Define the model
model = Sequential()
model.add(Dense(units=1, input_shape=(1,), activation='linear'))

# Compile the model
learning_rate = 0.02
optimizer = SGD(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Convert training data to NumPy arrays
x_train = np.array([1, 2, 3, 4])
y_train = np.array([2, 4, 6, 8])

# Train the model
epochs = 1000
batch_size = 2
model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size)
```

```python
import numpy as np
import matplotlib.pyplot as plt

model = Sequential()
model.add(Dense(units=1, input_shape=(1,)))

model.compile(optimizer='sgd', loss='mean_absolute_error' , metrics=['mae','mape'])

x_train = np.array([1, 2, 3, 4])
y_train = np.array([2, 4, 6, 8])

epochs = 1000
batch_size = 2
history = model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)
loss_history = history.history['loss']

plt.plot(range(epochs), loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

```python
loss_history = history.history['mae']


plt.plot(range(epochs), loss_history)
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.title('Training MAE')
plt.show()
```
